# Plotagem de gráficos

Vamos ver o básico de plotagem em _Julia_. 

Há várias maneiras diferentes de plotar um gráfico em _Julia_, incluindo chamar o _PyPlot_ ou o _Matplotlib_.

Aqui, usaremos o pacote nativo _Plots.jl_. Ele não está instalado, então precisaremos instalá-lo e ordenar que a linguagem precompile o mesmo.

```julia
using Pkg
Pkg.add("Plots")
using Plots
```

Uma das vantagens de usarmos o _Plots.jl_ é que ele nos permite mudar o backend perfeitamente. Aqui, vamos testar os backends _gr()_ e _plotlyjs()_.

Em nome da ciência, vamos usar nossas habilidades de programadores em _Julia_ para investigar a relação entre a temperatura global e o número de piratas desde 1860 até 200.

```julia
globaltemperatura = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpiratas = [45000, 20000, 15000, 5000, 400, 17];
```

Plots suportam diversos backends - isso é, bibliotecas que fazem o desenho em si - todos usando a mesma API. Para começar, vamos usar o backend GR. Escolhemos ele chamando

```julia
gr()
```

e usando comandos como _plo_ e _scatter_ para gerar os plots.

```julia
#Plotamos
plot(numpiratas, globaltemperatura, label="line")
```

```julia
# Plotamos marcando os pontos
plot(numpiratas, globaltemperatura, label="line")
scatter!(numpiratas, globaltemperatura, label="points")
```

O _!_ na função _scatter_ faz dela uma função mutável, indicando que os pontos destacados serão adicionados no plot pré-existente.

Agora vamos plotar com o nome dos eixos e título.

```julia
xlabel!("Numéro de piratas [aprox]")
ylabel!("Temperatura Global (C)")
title!("Influência da população de piratas no aquecimento global")
```

Isso ainda não parece muito legal! O número de piratas caiu desde 1860, então ler o plot da esquerda para a direita é meio que olhar para trás no tempo em vez de para frente. Vamos fazer um _flip_ no eixo x para vermos melhor esse gráfico.

```julia
xflip!()
```
